##### Preprocessing our data

###### converting catergorical variables into discrete numbers

```python
# here we use label encoder to convert categorical to numerical
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
## training data
df_train[1]=le.fit_transform(df_train[1])
df_train[2]=le.fit_transform(df_train[2])
df_train[3]=le.fit_transform(df_train[3])
df_train[41]=le.fit_transform(df_train[41])
```

```python
# testing data
df_test[1]=le.fit_transform(df_test[1])
df_test[2]=le.fit_transform(df_test[2])
df_test[3]=le.fit_transform(df_test[3])
df_test[41]=le.fit_transform(df_test[41])
```

```python
df_train.head()
```

```python
df_test.head()
```

```python
from sklearn.ensemble import RandomForestClassifier
model=RandomForestClassifier(n_estimators=100)
```

```python
x_train=df_train.iloc[:,df_train.columns !=41]
y_train=df_train.iloc[:,41]
```

```python
model.fit(x_train,y_train)
```

```python
x_test=df_test.iloc[:,df_test.columns !=41]
y_test=df_test.iloc[:,41]
```

```python
y_predict=model.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix
```

```python
accuracy_score(y_test,y_predict)
```

```python
plt.figure(figsize=(10,6))
sns.heatmap(confusion_matrix(y_test,y_predict),annot=True,fmt='d')
```
